```python
import pandas as pd
import numpy as np
import json
```

```python
df = pd.read_csv('cellCoords_0.tsv', delimiter='\t')
```

```python
df
```

```python
df['coords'] = [json.loads(df.coords[n]) for n in range(df.shape[0])]
df.to_json('cellCoords.json', orient='records')
```

```python
df = pd.read_csv('../cellData/cellData_0.tsv', delimiter='\t')
df["ClassName"] = '["Generic"]'
df['Genenames'] = [json.loads(df.Genenames[n]) for n in range(df.shape[0])]
df['CellGeneCount'] = [json.loads(df.CellGeneCount[n]) for n in range(df.shape[0])]
df['ClassName'] = [json.loads(df.ClassName[n]) for n in range(df.shape[0])]
df['Prob'] = [json.loads(df.Prob[n]) for n in range(df.shape[0])]
df.to_json('../cellData/cellData.json', orient='records')
```

```python
df = pd.read_csv('../geneData/geneData_0.tsv', delimiter='\t')
df['neighbour_array'] = [json.loads(df.neighbour_array[n]) for n in range(df.shape[0])]
df['neighbour_prob'] = [json.loads(df.neighbour_prob[n]) for n in range(df.shape[0])]

df.to_json('../geneData/geneData.json', orient='records')
```
